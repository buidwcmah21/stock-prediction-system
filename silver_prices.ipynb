```python
%sql
-- Tạo bảng Silver với tên chuẩn theo kế hoạch
CREATE TABLE IF NOT EXISTS workspace.stock_db.silver_prices (
  symbol STRING,
  date DATE,
  open DOUBLE,
  high DOUBLE,
  low DOUBLE,
  close DOUBLE,
  volume LONG,
  adj_close DOUBLE,
  price_change DOUBLE, 
  is_positive BOOLEAN, 
  processed_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (symbol);
```
