```python
%sql
-- Tạo bảng Gold chứa các chỉ số phân tích tài chính
CREATE TABLE IF NOT EXISTS workspace.stock_db.gold_indicators (
  symbol STRING,
  date DATE,
  close DOUBLE,
  daily_return_pct DOUBLE, 
  sma_20 DOUBLE,           
  sma_50 DOUBLE,           
  volatility_20d DOUBLE,   
  calculated_at TIMESTAMP
)
USING DELTA
PARTITIONED BY (symbol);
```
